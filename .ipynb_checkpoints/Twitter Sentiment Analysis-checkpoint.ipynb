##### Tokenization (Count vectorizer)

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

sample_data = ['This is the first paper.', 'This paper is the second paper.', 'And this is the third one.', 'Is this the first paper?']

X = vectorizer.fit_transform(sample_data)
```

```python
print(vectorizer.get_feature_names())
```

```python
print(X.toarray())
```

```python
mini_challenge_7 = ['Hello World', 'Hello Hello World', 'Hello World world worlD']
Y = vectorizer.transform(mini_challenge_7)

print(vectorizer.get_feature_names())
```

```python
print(Y.toarray())
```

```python
# Convert raw frequency counts into TF-IDF (Term Frequency -- Inverse Document Frequency) values
from sklearn.feature_extraction.text import TfidfTransformer
vectorizerTfmer = TfidfTransformer()

# Again, fit and transform
X_tfidf = vectorizerTfmer.fit_transform(X)
```

```python
# TF-IDF values
# raw counts have been normalized against document length, 
# terms that are found across many docs are weighted down ('the vs. 'second')

X_tfidf.toarray()
```

### Back to real data : Tweets

###### Creating a pipeline to remove punctuation, stopwords, and perform count vectorization

```python
#function to remove punctuation and stopwords

def message_cleaning(message):
    test_punc_removed ="".join([char for char in message if char not in string.punctuation])
    Test_pun_rem_clean =[word for word in test_punc_removed.split() if word.lower() not in stopwords.words('english')]
    return Test_pun_rem_clean
```

```python
tweets_df_cleaned = tweets_df['tweet'].apply(message_cleaning)
```

```python
print(tweets_df_cleaned[5])  ## cleand up version
```

```python
print(tweets_df['tweet'][5]) ## original version
```

```python
# initialize CountVectorizer
# fit and tranform using dataset

tweets_vectorizer= CountVectorizer(analyzer=message_cleaning, max_df = 0.8)
tweets_countvectorizer = tweets_vectorizer.fit_transform(tweets_df['tweet'])
```

```python
# huge dimensions! 31,962 documents, 47,386 unique terms.
tweets_countvectorizer.shape
```

```python
# Convert raw frequency counts into TF-IDF values
tweetsTfmer = TfidfTransformer()
tweets_tf_idf = tweetsTfmer.fit_transform(tweets_countvectorizer)
```

```python
# Same dimensions, now with tf-idf values instead of raw frequency counts
tweets_tf_idf.shape
```

```python
X = tweets_tf_idf
```

```python
Y = tweets_df['label']
```

```python
Y.shape
```

### Training Naive Bayes classifier

```python
#Split data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3 )
```

```python
# Now ready to build a classifier and train it 
# We will use Multinominal Naive Bayes as our model

from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

###### Assess Trained Model Performance

```python
import sklearn
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#Predicting the test set results, finding accuracy
y_predict_test = NB_classifier.predict(X_test)
sklearn.metrics.accuracy_score(y_test, y_predict_test)
```

```python
#Checking the confusion matrix
cm = confusion_matrix(y_test, y_predict_test)
cm
```

```python
#Visualizing confusion matrix
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict_test))
```

### Training Random Forest Classifier


```python
from sklearn.ensemble import RandomForestClassifier
text_classifier = RandomForestClassifier(n_estimators=200, random_state=0)
text_classifier.fit(X_train,y_train)
```

### Assess RF model Performance

```python
prediction_RF = text_classifier.predict(X_test)
```

```python
print(confusion_matrix(y_test,prediction_RF))
print(classification_report(y_test,prediction_RF))
print(accuracy_score(y_test, prediction_RF))
```
